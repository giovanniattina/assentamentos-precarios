```python
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import shapiro, anderson
```

```python
cnefe_hex_gpd = gpd.read_file("data/shapeFiles/sad69_hex_sp_cnefe_agg")
```

```python
cnefe_hex_gpd.head()
```

```python
# Filter for numerical columns

numerical_columns = cnefe_hex_gpd.drop(['fid', 'id_hex', 'abbrev_mun', 'name_muni', 'code_muni', 'geometry'], axis=1).select_dtypes(include=['number']).columns
```

```python

# Set up figure and axis
fig, axes = plt.subplots(nrows=len(numerical_columns), figsize=(10, 5 * len(numerical_columns)))

# Plot choropleth map for each numerical column
for i, column in enumerate(numerical_columns):
    ax = axes[i] if len(numerical_columns) > 1 else axes
    cnefe_hex_gpd.plot(column=column, cmap='OrRd', legend=True, ax=ax)
    ax.set_title(f"Choropleth Map of {column}")

# Show the plots
plt.tight_layout()
plt.show()
```

```python


# Create a figure for boxplots
fig, axes = plt.subplots(nrows=len(numerical_columns), figsize=(10, 5 * len(numerical_columns)))

# Plot boxplot and test normality for each variable
for i, column in enumerate(numerical_columns):
    ax = axes[i] if len(numerical_columns) > 1 else axes
    sns.boxplot(data=cnefe_hex_gpd, x=column, ax=ax)
    ax.set_title(f"Boxplot of {column}")

    # Perform Shapiro-Wilk normality test
    stat, p_value = shapiro(cnefe_hex_gpd[column].dropna())
    
    # Print result of normality test
    if p_value > 0.05:
        print(f"{column}: Likely Normally Distributed (p-value = {p_value:.3f})")
    else:
        print(f"{column}: Not Normally Distributed (p-value = {p_value:.3f})")
        
    # Perform Anderson-Darling test (optional, stricter test)
    anderson_result = anderson(cnefe_hex_gpd[column].dropna())
    print(f"Anderson-Darling test statistic for {column}: {anderson_result.statistic:.3f}")

# Show the plots
plt.tight_layout()
plt.show()
```

```python
correlation_matrix = cnefe_hex_gpd.drop(['fid', 'id_hex', 'abbrev_mun', 'name_muni', 'code_muni', 'geometry'], axis=1).corr()

sns.heatmap(correlation_matrix)
```
