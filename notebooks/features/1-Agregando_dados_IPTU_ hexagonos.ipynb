```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
DEFAULT_CRS = "EPSG:4326"
```

```python

hex_sp = gpd.read_file("data/shapeFiles/GRIDS_H3_SP_RES10") 


# Transform the CRS to EPSG:31983
hex_sp = hex_sp.to_crs(DEFAULT_CRS)
hex_sp.head()
```

# IPTU Lotes

```python
iptu_lotes = gpd.read_parquet('data/raw/iptu_por_lotes_2020.parquet')
```

```python
iptu_lotes.crs
```

```python
iptu_lotes.head()
```

```python
print(list(iptu_lotes.columns))
```

```python
iptu_lotes[iptu_lotes['sqlc'] == '170041007100']
```

```python
iptu_lotes['qnt_unidades_residencias'] = iptu_lotes[[item for item in list(iptu_lotes.columns) if ('residencial' in item.lower() and 'qt' in item.lower())]].sum(axis=1)
```

```python
iptu_lotes_final = iptu_lotes[['sqlc', 'qnt_unidades_residencias', 'geometry']]
```

```python
iptu_lotes_final['qnt_unidades_residencial_m2'] = iptu_lotes_final['qnt_unidades_residencias'] / (iptu_lotes_final.to_crs("EPSG:5641").area)
```

```python
iptu_lotes_final.head()
```

# Intersecção

```python
if iptu_lotes_final.crs != DEFAULT_CRS:
    iptu_lotes_final = iptu_lotes_final.to_crs(DEFAULT_CRS)
if hex_sp.crs != DEFAULT_CRS:
    hex_sp = hex_sp.to_crs(DEFAULT_CRS)
```

```python
iptu_lotes_join_h3 =  iptu_lotes_final.overlay(hex_sp[['id_hex', 'geometry']], how='intersection')
```

```python
iptu_lotes_join_h3.head()
```

```python
iptu_lotes_join_h3[iptu_lotes_join_h3['sqlc'] == '005076017300']
```

## Krigatem simples e Agrupar 

```python
iptu_lotes_join_h3['qnt_unidades_residencias']  = iptu_lotes_join_h3['qnt_unidades_residencial_m2'] * (iptu_lotes_join_h3.to_crs("EPSG:5641").area)
```

```python
iptu_lotes_join_h3.head()
```

```python
iptu_lotes_join_h3[iptu_lotes_join_h3['sqlc'] == '170041007100']
```

```python
iptu_lotes_agg = iptu_lotes_join_h3.groupby(['id_hex']).agg({
    "qnt_unidades_residencias": ['sum', 'mean', 'median', 'std', 'var'],
}).reset_index()
```

```python
iptu_lotes_agg[iptu_lotes_agg['id_hex'] == '8aa81005086ffff']
```

```python
iptu_lotes_agg.columns = ['_'.join(col).strip() for col in iptu_lotes_agg.columns.values if col != 'id_hex']
```

```python
iptu_lotes_agg.head()
```

```python
iptu_lotes_agg[iptu_lotes_agg['id_hex_'] == '8aa8100c0747fff']
```

```python
hex_iptu = hex_sp.join(iptu_lotes_agg.set_index('id_hex_'), how='left', on='id_hex')
```

```python
hex_iptu["qnt_unidades_residencias_sum"] = np.ceil(hex_iptu["qnt_unidades_residencias_sum"])
```

```python
hex_iptu.head()
```

```python
hex_iptu.fillna(0, inplace=True)
```

```python
hex_iptu.isna().sum()
```

```python

# Write the GeoDataFrame to a shapefile
hex_iptu.to_file("data/generate/WGS84_hex_iptu_2020.gpkg", driver="GPKG")
```
