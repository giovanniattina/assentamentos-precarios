```python
import geopandas as gpd
```

```python
DEFAULT_CRS = "EPSG:4326"
```

# lendos shapefiles



```python
hex_iptu = gpd.read_file("data/shapeFiles/WGS84_hex_iptu")
hex_contruco = gpd.read_file("data/WGS84_hex_construcao_open_buildings_2023.gpkg")
hex_cnefe = gpd.read_file("data/shapeFiles/hex_agg_cnef_sp", columns=['fid', 'esp_1', 'esp_sum'])
hex_census = gpd.read_file("data/WGS84_hex_census_2010.gpkg")
```

```python
hex_census.head()
```

```python
def set_default_crs(gdf):
    return gdf.to_crs(DEFAULT_CRS)
```

```python
hex_iptu = set_default_crs(hex_iptu)
hex_contruco = set_default_crs(hex_contruco)
hex_cnefe = set_default_crs(hex_cnefe)
hex_census = set_default_crs(hex_census)
```

# Agrupando arquivos

```python

hex_f = hex_iptu.join(hex_contruco[['hex_fid', 'area_sum', 'area_mean', 'area_median', 'area_std', 'area_var', 'area_count', 'area_const']].set_index("hex_fid"), how='left', on='fid')
hex_f = hex_f.join(hex_cnefe[['fid', 'esp_1', 'esp_sum']].set_index("fid"), how='left', on='fid')
hex_f = hex_f.join(hex_census[['hex_fid', 'D1_012_d_sum', 'D1_017_d_sum', 'D1_035_d_sum', 'D1_044_d_sum']].set_index("hex_fid"), how='left', on='fid')


```

```python
hex_f.head()
```

# Transformações finais
- fill all na

```python
hex_f.fillna(0, inplace=True)
```

```python
hex_f.to_file("data/shapeFiles/WGS84_hex_agg", driver='ESRI Shapefile')
```
